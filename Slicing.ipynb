What is Slicing in Python?

Slicing is a technique used to extract a portion (subsequence) of elements from a sequence data type such as:

list

tuple

string

String Slicing

```python
text = "Mohsin Ali will become a Machine Learning Engineer One Day. Inshallah!"
print(text[1:7])
print(text[1:25])
print(text[0:7])
print(text[:12])
print(text[2:])
print(text[-8:])     #Reverse the string
print(text[::2])     #Every second character
```

List Slicing

```python
numbers = [10,20,30,40,50,60,70,80,90]
print(numbers[1:2])
print(numbers[2:])
print(numbers[:2])
print(numbers[-1:])
print(numbers[:-1])
print(numbers[::2])
print(numbers[::-1])
```

Tuple Slicing

```python
tuple = (10,20,30,40)
print(tuple[1:2])
print(tuple[2:1])
print(tuple[::-1])
print(tuple[-1::])
print(tuple[::-1])
```

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

print(letters[::3])    
print(letters[1::2])     
```

```python
#modifying list using slicing
list = [1,5,6,7,8]
list[1:] = [2,3,4]
print(list)
```

```python
#delete using slicing
nums = [1,2,3,4,5]
del nums[1:3]
print(nums)
```

```python
x = "PythonProgramming"
print(x[2:10:2])
```

```python
#reverse list
list = [1,2,3,4,5]
print(list[::-1])
```

```python
#extract last three characters of a string

st = "Mohsinali"
str = st[-3:]
print(str)
```
