```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        avg = sum(self.marks)/ len(self.marks)

        if avg>=80:
            return "A"
        elif avg>=60:
            return "B"
        elif avg>=50:
            return "C"
        else:
            return "Fail"
    
    def display(self):
        print(f"\nName : {self.name}")
        print(f"Marks : {self.marks}")
        print(f"Grade : {self.calculate_grade}")

    
    
def add_student(student_list):
    try:
        name = input("Enter student name:")

        marks = []
        for i in range(3):
            mark = int(input(f"Enter mark {i+1}:"))
            marks.append(mark)
        
        student = Student(name,marks)
        student_list.append(student)

        print("Student added sucessfully!")



    except ValueError:
        print("Invalid input")


def show_students(student_list):
    if not student_list:
        print("No student found")
        return
    print("\n-----Student List-----")
    for student in student_list:
        student.display()


def find_topper(student_list):
    try:
        if not student_list:
            print("No student Available")
            return
        top_student = student_list[0]

        for i in student_list:
            avg1 =sum(i.marks)/len(i.marks)
            avg2 = sum(top_student.marks)/len(top_student.marks)

            if avg1>avg2:
                top_student = i

        print("\n Topper is")
        top_student.display()

    except Exception as e:
        print("Error occurred", e)

def main():
    students = []
    while True:
        print("\n========Menu========")
        print("1. Add Student")
        print("2. Show Student")
        print("3. Find Topper")
        print("4. Exit")

        try:
            choice = int(input("Enter choice betwee 1-4"))

            if choice ==1:
                add_student(students)
            elif choice ==2:
                show_students(students)
            elif choice ==3:
                find_topper(students)

            elif choice ==4:
                print("Exiting Program")
                break
            else:
                print("Invalid choice")

        except ValueError:
            print("Please Enter a valid number")


main()
```
