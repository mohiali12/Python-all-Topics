Inheritance:



super(): super() Python me parent class ke methods ya constructor ko call karne ke liye use hota ha

```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def driving(self):
        print("i can drive this car")

# Audi car is inherting from car class
class Audi(Car):
    def __init__(self,windows,doors,enginetype,horsepower): # add new feature in the child class "horsepower"
        super().__init__(windows, doors, enginetype)     #inheriting from parent class
        self.horsepower = horsepower


a1 = Audi(5,4,"V8",600)
print(a1.horsepower)
print(a1.enginetype)
print(a1.doors)
print(a1.windows)
a1.driving()

```

```python
class Person:
    def __init__(self, name , age , gender):
        self.name = name 
        self.age = age
        self.gender = gender
        print("Person constructor executed")

    def display_Person_info(self):
        print("\n------Person Information-------")
        print("Name :", self.name)
        print("Age :", self.age)
        print("Gender :", self.gender)


    def get_role(self):
        return "General Person"
    


# Derived Class

class Employee(Person):
    def __init__(self, name , age , gender , emp_id, salary):
        super().__init__(name, age , gender)
        self.emp_id = emp_id
        self.salary = salary
        print("Employee constructor executed")

    def display_employee_info(self):
        print("\n--------Employee information--------")
        print("Employee ID :", self.emp_id)
        print("Salary :", self.salary)

    def calculate_annual_salary(self):
        return self.salary*12
    
    def get_role(self):
        return "Employee"
    



class Manager(Employee):
    def __init__(self, name , age , gender , emp_id , salary , department, bounus ):
        super().__init__(name, age , gender , emp_id, salary)
        self.department = department
        self.bonus = bounus
        print("Manager Constructor executed")

    def display_manager_info(self):
        print("\n--------Manager information---------")
        print("Department :", self.department)
        print("Bonus :", self.bonus)

    def calculate_total_income(self):
        annual_salary = self.calculate_annual_salary()
        return annual_salary + self.bonus
    
    def get_role(self):
        return "Manager"
    


# now we will create object

print("\n Creating Manager Object....")

manager1 = Manager(
    "Mohsin",
     23,
     "Male",
    "1123",
    80000,
    "IT",
    20000
)

manager1.display_Person_info()
manager1.display_employee_info()
manager1.display_manager_info()


print("\n Role: ", manager1.get_role())

print("\n Annual Salary :", manager1.calculate_annual_salary())
print("\n Total income :", manager1.calculate_total_income())





```

Encapsulation:
is an Object-Oriented Programming principle where:

Data (variables/attributes) and methods (functions) are bundled inside a class, and direct access to some data is restricted.

It is mainly used for:

Data protection
Data hiding
Controlled access
Improving security & maintainability

```python
class A:
    _a = 2    #protected (use outside the class)
    __b = 3   #private (not use outside the class)

    def show(self):
        print("value of a is", self._a)
        print("Vlaue of b is", self.__b)

# this can be print because , show is the function of A class. 
obj = A()
obj.show()


# otherwise if we want to print value outside the classit will give error

# secode line will give error becasue it can find the attribute __b.

print("outside the class", obj._a)
#print("outside the class", obj.__b)
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self._balance = balance                   # protected

    def _validate_amount(self,amount):
        return amount>0
    
    def _update_balance(self,amount):
        self._balance += amount

    def deposit(self,amount):
        if self._validate_amount(amount):
            self._update_balance(amount)
            print("Deposit Successful")

    def withdraw(self, amount):
        if self._validate_amount(amount) and self._balance >=amount:
            self._update_balance(-amount)
            print("Withdraw successful")
        else:
            print("Insufficient balance")

    def show_balance(self):
        print(f"Current Balacne :{self._balance}")


class SavingAccount(BankAccount):
    def add_interest(self):
        interest = self._balance *0.05
        self._update_balance(interest)
        print("Interest added")


acc = SavingAccount("Mohi",10000)
acc.deposit(2000)
acc.withdraw(1000)
acc.add_interest()
acc.show_balance()

    
        
```
