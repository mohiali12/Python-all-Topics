```python
plt.plot(data['index'], data['V Kohli'])
plt.plot(data['index'], data['RG Sharma'])
```

Topic: Colors , Line(width and style), marker(size)

```python
plt.title('Virat kohli and Rohit Sharma Comparison')
plt.xlabel('Year')
plt.ylabel('Score')
plt.plot(data['index'], data['V Kohli'], color = 'Purple', linestyle = 'solid', linewidth = 6, marker = 'D')
plt.plot(data['index'], data['RG Sharma'], color= 'green', linestyle = 'dashdot', marker = 'o', markersize = 10)

# line style has three style .... dotted , dahsed dashdot, and solid 

# marker is use to point in shape like belowin code , it can be (D, +, ., >, <, O)
```

```python
plt.title('Virat kohli and Rohit Sharma Comparison')
plt.xlabel('Year')
plt.ylabel('Score')
plt.plot(data['index'], data['V Kohli'], color = "green", label = 'Virat')
plt.plot(data['index'], data['RG Sharma'], color = 'purple', label = 'Rohit')

plt.legend()

# here legend is use for show line names , as label
```

```python
plt.title('Virat kohli and Rohit Sharma Comparison')
plt.xlabel('Year')
plt.ylabel('Score')
plt.plot(data['index'], data['V Kohli'], color = "green", label = 'Virat')
plt.plot(data['index'], data['RG Sharma'], color = 'purple', label = 'Rohit')

plt.legend(loc = 'lower right')
plt.grid()

# and if we want to change location of showing name in graphs so we can use "loc"

# grid show line in background
```
