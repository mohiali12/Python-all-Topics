Handling missing Values

```python
# NaN (not a number)
# None
# isnull()
# True = value is missing
# False = value is present



# we will make a new dataset for check missing values
```

```python
import pandas as pd
import numpy as np
data={
    "Name":['Mohi',"wICKI",'Ali','Zebi','Viki','Joni','Ahmed'],
    "Age":[23,None,21,27,26,31,25],
    "Salary":[50000,None,70000,80000,450000,55000,75000],
    "Performance":[85,None,79,89,91,93,81]
}

df = pd.DataFrame(data)
print(df)

print(df.isnull())
```

Dropna(): if any row have even one nan value , the full row will be remove

```python
# if we want to remove missing values from dataset 

''' we have major difference see here , when i don't put inplace = True , it
give copy of dataset . it does't change into original datset
, as you see we have one emlement is more in name "wICKI"  so when run
the code 1 rows become more and other vlaues are NaN,'''

''' if we put inplace = True then the wICKI element will be reomve from datset 
because inplace = True means it change original datset 
'''
''' also the concept of dropna is that if a NaN value exist in a row then the 
full row will be remove'''

df.dropna()
print(df)
```

```python
df.dropna(inplace=True)
print(df)
```

```python
# add value with fillna like below in code
import pandas as pd


data= {
    "Name": [None,"Ali","Roshni","Mohi","Shamme","Bablu"],
    "Age":[21,None,23,24,18,20],
    "Salary":[20000,30000,None,60000,70000,80000]
}
df = pd.DataFrame(data)

df.fillna(0, inplace=True)
print(df)
```

```python
import pandas as pd


data= {
    "Name": [None,"Ali","Roshni","Mohi","Shamme","Bablu"],
    "Age":[21,None,23,24,18,20],
    "Salary":[20000,30000,None,60000,70000,80000]
}
df = pd.DataFrame(data)

df["Age"].fillna(df["Age"].mean(), inplace=True)
df['Salary'].fillna(df["Salary"].mean(), inplace=True)
print(df)
```
